```python
import random

def duration_with_vc(mean_duration, vc=0.1):
    std_dev = mean_duration * vc
    return max(0.1, random.gauss(mean_duration, std_dev))  # kein negativer Wert

        
```

```python
for i in range (0, 50):
    print(duration_with_vc(29, vc=0.1))
```

```python
def duration_lognormal(mean_duration, vc=0.2):
    """
    Liefert eine realistische, lognormal-verteilte Dauer:
    - Erwartungswert liegt etwa bei `mean_duration`
    - VC = Variationskoeffizient (z.â€¯B. 0.2 fÃ¼r 20â€¯% Streuung)
    """
    sigma = math.sqrt(math.log(1 + vc**2))
    mu = math.log(mean_duration) - 0.5 * sigma**2
    return random.lognormvariate(mu, sigma)
```

```python
liste = []
for i in range (0, 100):
    val = duration_lognormal(30, vc=0.2)
    print(val)
    liste.append(val)
```

```python
import random
import math

def duration_lognormal(duration, vc=0.2):
    """
    Liefert eine Lognormal-verteilte Zufallsdauer basierend auf:
    - duration: gewÃ¼nschter (aber nicht garantierter) Mittelwert
    - vc: Variationskoeffizient (z.â€¯B. 0.2 fÃ¼r 20 %)
    Achtung: Der resultierende Erwartungswert liegt **oberhalb** von duration.
    """
    sigma = vc  # direkt verwenden
    mu = math.log(duration)
    return random.lognormvariate(mu, sigma)
```

```python
liste = []
for i in range (0, 100):
    val = duration_lognormal(30, vc=0.2)
    liste.append(val)

print(f"Max {max(liste)}")
print(f"Min {min(liste)}")
print(f"Mean {sum(liste) / len(liste)}")
```
